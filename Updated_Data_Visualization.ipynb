```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./group12/clean_data.csv')
```

```python
# Visualization 1: Stacked Bar Chart of Opinions on Gun Control Laws by Party Affiliation
gunlaw_party = df.groupby(['gunlaw', 'partyid']).size().unstack()
gunlaw_party_percentage = gunlaw_party.div(gunlaw_party.sum(axis=1), axis=0)
subplot = gunlaw_party_percentage.plot(kind='bar', stacked=True)

for i in subplot.patches:
    width, height = i.get_width(), i.get_height()
    x, y = i.get_xy()
    percentage = f'{height * 100:.2f}%'
    subplot.annotate(percentage, (x + width / 2, y + height / 2), ha='center', va='center',
                fontsize = 8)

plt.title('Opinions on Gun Control Laws by Party Affiliation')
plt.xlabel('Gun Control Laws Opinion')
plt.ylabel('Proportion')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.xticks(rotation=0)
plt.show()

```

```python
# Visualization 2: Grouped Bar Chart of Opinions on Gun Laws by Race
pornlaw_party = df.groupby(['race', 'gunlaw']).size().unstack()
pornlaw_party.plot(kind='bar')
plt.title('Opinions on Gun Laws by Race')
plt.xlabel('Gun Control Laws Opinion')
plt.ylabel('Count')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Visualization 3: Histograms of Years of Education and Opinion on Gun Laws

opinions = df['gunlaw'].unique()
plt.figure(figsize=(12, 10))

xlimits = (0, 20)
xticks = list(range(0, 22, 2))
for i, o in enumerate(opinions):
    plt.subplot(1, 3, i + 1)
    subset = df[df['gunlaw'] == o]
    plt.hist(subset['educ'], bins=10, alpha=0.7)
    plt.xlim(xlimits)
    plt.xticks(xticks)
    plt.title(f' Years of Education on Gun Laws\n({o})')
plt.tight_layout()
plt.show()


```
